```python
file = open(r".../input/rosalind_revc.txt")
dna = file.read()
print(dna)
```

```python
rev = ''
for i in dna:
    if i =="A":
        rev = "T" + rev
    elif i =="T":
        rev = "A" + rev
    elif i =="G":
        rev = "C"+ rev
    elif i =="C":
        rev = "G" +rev
print(rev)
        
```
